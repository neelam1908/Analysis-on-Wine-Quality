# **4. Data Acquisition & Description**
---

- This section is emphasised on the accquiring the data and obtain some descriptive information out of it.

- You could either scrap the data and then continue, or use a direct source of link (generally preferred in most cases).

- Before going further you must have a good idea about the features of the data set:

|Id|Feature|Description|
|:--|:--|:--|
|01||fixed acidity| most acids involved with wine or fixed or nonvolatile (do not evaporate readily).| 
|02|volatile acidity|the amount of acetic acid in wine, which at too high of levels can lead to an unpleasant, vinegar taste.| 
|03|citric acid|found in small quantities, citric acid can add ‘freshness’ and flavor to wines.| 
|04|residual sugar | the amount of sugar remaining after fermentation stops, it’s rare to find wines with less than 1 gram/liter |
|||and wines with greater than 45 grams/liter are considered sweet.|
|05| chlorides| the amount of salt in the wine.|
|06| free sulfur dioxide  | free form of SO2 exists in equilibrium between molecular SO2 (as a dissolved gas) and bisulfite ion;|
|||it prevents microbial growth and the oxidation of wine.|
|07|  total sulfur dioxide | amount of free and bound forms of S02; in low concentrations, SO2 is mostly undetectable in wine,|
|||but at free SO2 concentrations over 50 ppm, SO2 becomes evident in the nose and taste of wine|
|08| density| the density of water is close to that of water depending on the percent alcohol and sugar content|
|09| pH | describes how acidic or basic a wine is on a scale from 0 (very acidic) to 14 (very basic);|
||| most wines are between 3-4 on the pH scale.|
|10| sulphates| a wine additive which can contribute to sulfur dioxide gas (S02) levels, wich acts as an antimicrobial and antioxidant.|
|11| alcohol| the percent alcohol content of the wine.|
|12| quality| score between 0 to 10.|

```python
df_wine = pd.DataFrame()    #defining an empty dataframe 
#assigning column heads and data with respective to original dataset
df_wine['density'] = df_winequal['density']
df_wine['alcohol'] = df_winequal['alcohol']
df_wine['quality'] = df_winequal['quality']
df_wine['salty'] = df_winequal['chlorides']
df_wine['sweetness'] = df_winequal['residual sugar']
df_wine['sourness'] = df_winequal['fixed acidity'] + df_winequal['volatile acidity'] + df_winequal['citric acid'] + df_winequal['pH']
df_wine['odour'] = df_winequal['sulphates'] + (df_winequal['free sulfur dioxide']/1000) + (df_winequal['total sulfur dioxide']/1000)
df_wine.head()
```

```python
df_wine.describe()
```

```python
sns.pairplot(data= df_wine, hue = 'quality', diag_kind='hist', height = 1.3)
```

# **9. Summarization**
---

```python
plt.figure(figsize= (15,8))
plt.title('Stacked Barchart for Wine Quality')
df_wine.groupby(['quality'])['sourness'].mean().plot(kind='bar', color='red', stacked=True)
df_wine.groupby(['quality'])['alcohol'].mean().plot(kind='bar', color='blue', stacked=True)
df_wine.groupby(['quality'])['sweetness'].mean().plot(kind='bar', color='green', stacked=True)
df_wine.groupby(['quality'])['density'].mean().plot(kind='bar', color='orange', stacked=True)
df_wine.groupby(['quality'])['odour'].mean().plot(kind='bar', color='brown', stacked=True)
df_wine.groupby(['quality'])['salty'].mean().plot(kind='bar', color='black', stacked=True)
plt.legend(['sourness','alcohol','sweetness','density','odour','salty'], bbox_to_anchor=(1,1))
plt.yticks(np.arange(0,14,1))
```

### **9.1 Conclusion**

- Sourness is a dominant character in low to medium quality wines while alcohol is a dominant character for high quality wines.

- Salty & Odour constitute a minute role and individually may have less impact on quality.

- Sweetness level proves to be an important factor across all the quality scores.

<a name = Section92></a>
### **9.2 Actionable Insights**

- Higher the alcohol content, higher is the quality score. Focus should be alcohol content.
- Decrease in the sourness and medium to high increase in sweetness for good quality wine.
- Salty and Odour remain consistent and minimum for good quality. 
- Density seems to be a consistent actor across all the samples.
